```python
def display_feature(data, regions, feature):
    
    for region in regions:
        to_be_plotted = data.loc[(data['Indicator Name'] == feature) & (data['Country Name'] == region)]
        plt.plot(to_be_plotted.iloc[0, 3:], label=region)
        plt.title(feature)
    plt.legend()
    plt.show()
```

```python
display_feature(df_regions, regions_selected, 'GDP (constant 2010 US$)')
```

```python
display_feature(df_regions, regions_selected, 'Urban population (% of total population)')
```

```python
display_feature(df_regions, regions_selected, 'School enrollment, primary (% gross)')
```

```python
# 'Land area (sq. km)'
display_feature(df_regions, regions_selected, 'Land area (sq. km)')
```
