```python
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px
import pandas as pd
from sklearn.decomposition import PCA
from sklearn import preprocessing
import numpy as np
```

```python
df = pd.read_csv('preprocessed_data.csv')
df.head()
```

```python
indicators = ['GDP',
             'CPI (consumer price index)',
             'Real interest rate / deposit interest rate',
             'Adjusted net national income',
             'Export of goods and services',
             'Unemployment',
             'Life expectancy at birth',
             'school enrollment/ school attainment',
             'suicide mortality rate',
             'CPIA property rights and rule-based governance rating',
             'Population, total',
             'Mortality rate, under-5 (per 1,000 live births)',
             'Life expectancy at birth, total (years)',
             'Age dependency ratio (% of working-age population)',
             'Fertility rate, total (births per woman)',
             'Land area (sq. km)',
             'Agricultural land (sq. km)',
             'Urban population (% of total population)',
             'Population density (people per sq. km of land area)']
```

```python
years_list = list(df.columns)
years_list.pop(0)
years_list.pop(0)
years_list
```

```python
df = df.fillna(0)
df_pivot = pd.pivot_table(df, values=years_list, index='Country Name', columns='Indicator Name')
df_pivot['1990'].head()
```

```python
df_pivot.fillna(0)
for year in years_list:
    df_temp = df_pivot[year]
    if year == '1989':
        array = df_temp.values
    else:
        temp_array = df_temp.values
        array = np.concatenate((array, temp_array), axis=0)
    
array_scaled = preprocessing.scale(array)
pca = PCA()
pca.fit(array_scaled)
```

```python
app = dash.Dash(__name__)

app.layout = html.Div([
    dcc.Graph(id="scatter-plot"),
    html.P("Year:"),
    dcc.Slider(
        id='slider',
        min=1989, max=2019, step=1,
        marks={1989: '1989',
               1990: '1990',
               2000: '2000',
               2010: '2010',
               2019: '2019'},
        value=2019),
])

@app.callback(
    Output("scatter-plot", "figure"), 
    [Input("slider", "value")])
def update_bar_chart(slider):
    year = slider
    df = df_pivot[str(year)]
    array = df.values
    array_scaled = preprocessing.scale(array)
    X_pca = pca.transform(array_scaled)
    fig = px.scatter(
        x=X_pca[:, 0], y=X_pca[:, 1],
        text=list(df_pivot['1989'].index))
    return fig

app.run_server(debug=False)
```
