```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../data/features_selected.csv', index_col=0)
df.head(10)
```

# inspect lacking countries

```python
nan_values_in_countries = {}
total_country_entries = df[df['Country Name']=='Germany']['Country Name'].count()

for country in df['Country Name'].unique():
    nan_values_in_countries[country] = df[df['Country Name'] == country].isna().sum().sum()
   
print(nan_values_in_countries)
print(total_country_entries)
```

```python
regions = []
for country in df['Country Name'].unique():
    regions += [country]
    if country == 'World':
        break
    
# 47 pierwszych to regiony
regions

regions_df = df[df['Country Name'].isin(regions)]

regions_df
```

```python
countries_df = df[-df['Country Name'].isin(regions)]
countries_df
```

# finding nans throughout data

```python
print(f'regions nan count: {np.sum(regions_df.isna()).sum()}, \ncountries nan count: {np.sum(countries_df.isna()).sum()}')


df[df['Country Name'] == "Germany"].isna().sum().sum()
```

```python

def count_nans_per_country(df):
    countries_nans = {}
    for country_name in df['Country Name'].unique():
        countries_nans[country_name] = df[df['Country Name'] == country_name].isna().sum().sum()
    return countries_nans

countries_nans = count_nans_per_country(countries_df)
regions_nans = count_nans_per_country(regions_df)
```

```python
nan_values_throughout_countries = pd.DataFrame.from_dict(countries_nans, orient='index', columns=['nans'])
nan_values_throughout_countries
```

```python

135/ (len(countries_df['Indicator Name'].unique()) * len(countries_df.columns[2:]))
```

```python
nan_values_throughout_countries.sort_values('nans', ascending=False)
nan_values_throughout_countries.describe()
```

```python
first_dropout = nan_values_throughout_countries[nan_values_throughout_countries['nans'] < nan_values_throughout_countries['nans'].quantile(.75)]
first_dropout
```

```python
df_after_dropout = df[df['Country Name'].isin(first_dropout.index)]
df_after_dropout['Country Name'].unique().shape
```

# inspect lacking indexes

```python
nan_vals_throughout_indices = {}
total_number_of_cells_per_index = df_after_dropout['Indicator Name'].nunique() * (df.shape[1] - 2) * df_after_dropout['Country Name'].nunique()
print(total_number_of_cells_per_index)
for index in df_after_dropout['Indicator Name'].unique():
    nans_number = df_after_dropout[df_after_dropout['Indicator Name'] == index].isna().sum().sum()
    nan_vals_throughout_indices[index] = (nans_number, np.round(nans_number/total_number_of_cells_per_index, 4))
nan_vals_throughout_indices
```

```python

nan_values_of_indices = pd.DataFrame.from_dict(nan_vals_throughout_indices, orient='index', columns=['nans', 'nans ratio ind'])
nan_values_of_indices['total nans ratio'] = nan_values_of_indices['nans']/nan_values_of_indices['nans'].sum()


print(nan_values_of_indices.sort_values('nans', ascending=False))
print('\n\n', nan_values_of_indices.describe())

```

```python
second_dropout = nan_values_of_indices[nan_values_of_indices['nans ratio ind'] < .01]
second_dropout.shape
```

```python
df_after_2nd_dropout = df_after_dropout[df_after_dropout['Indicator Name'].isin(second_dropout.index)]
df_after_2nd_dropout.shape
```

```python
total = 0
for country in df_after_2nd_dropout['Country Name'].unique():
    for index in df_after_2nd_dropout['Indicator Name'].unique():
        if(df_after_2nd_dropout[(df_after_2nd_dropout['Country Name']==country) & (df_after_2nd_dropout['Indicator Name']==index)].isna().sum().sum() > 7):
            print(country, '||', index)
            total += 1
print(total)
```

```python
def counts_nans_in_row_above(data, n):
    countries = []
    total = 0
    for country in data['Country Name'].unique():
        for index in data['Indicator Name'].unique():
            if(data[(data['Country Name']==country) & (data['Indicator Name']==index)].isna().sum().sum() > n):
                countries += [country]
                print(country, '||', index)
                total += 1
    print(total)
    return countries
    
```
