```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('./data/non_nans.csv', index_col=0)
df.head()
```

```python
df['Indicator Name'].unique()
```

```python
def row_normalization(country, indicator, time_series):
    dict_temp = dict()
    dict_temp['Country Name'] = country
    normalization_values = df[df['Country Name']==country]
    
    if indicator == 'Adjusted net national income (constant 2010 US$)':
        normalization_values = normalization_values[ normalization_values['Indicator Name']=='Population, total'].values[0][2:]
        normalization_values /= 1000
    elif indicator == 'Agricultural land (sq. km)':
        normalization_values = normalization_values[ normalization_values['Indicator Name']=='Land area (sq. km)'].values[0][2:]
    elif indicator == 'Exports of goods and services (constant 2010 US$)':
        normalization_values = normalization_values[ normalization_values['Indicator Name']=='Population, total'].values[0][2:]
        normalization_values /= 1000
    elif indicator == 'Exports of goods and services (constant 2010 US$)':
        normalization_values = normalization_values[ normalization_values['Indicator Name']=='Population, total'].values[0][2:]
        normalization_values /= 1000
    elif indicator == 'GDP (constant 2010 US$)':
        normalization_values = normalization_values[ normalization_values['Indicator Name']=='Population, total'].values[0][2:]
        normalization_values /= 1000
    else:
        normalization_values = 1
    
    time_series = time_series/normalization_values
    dict_temp['Indicator Name'] = indicator
    
    columns = list(df.columns)
    columns = columns[2:]
    
    for year, value in zip(columns, time_series):
        dict_temp[year] = value
    
    return dict_temp
        
```

```python
list_normalized = list()
for row in df.iterrows():

    country = row[1][0]
    indicator = row[1][1]
    time_series = row[1][2:].values
    
    list_normalized.append(row_normalization(country, indicator, time_series))
    
    df_normalized = pd.DataFrame(list_normalized)
```
